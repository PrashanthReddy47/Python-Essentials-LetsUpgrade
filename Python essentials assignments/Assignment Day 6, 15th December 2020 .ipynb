```python
# mail and password using Mailtrap for safe email testing
# import components:
import smtplib
from socket import gaierror

# Define the SMTP server here:
port = 587
smtp_server = "smtp.mailtrap.io"
login = "jakegrim777@trap.com" # login generated by Mailtrap
password = "Kwerty@123" # password generated by Mailtrap

# sender’s and receiver’s email addresses:
sender = "jakegrim777@trap.com"
receiver = "grimjake777@trap.com"

#'f' used to  automatically insert variables in the text.
message = f"""\ 
Subject: Hi Mailtrap
To: {receiver}
From: {sender}
First message using Python.""" # Message

try:
  # Sending message with above credentials.
  with smtplib.SMTP(smtp_server, port) as server:
    server.login(login, password)
    server.sendmail(sender, receiver, message)
except (gaierror, ConnectionRefusedError):
  # If my message was sent or which errors need to be fixed then it will report
  print('Failed to connect to the server. Bad connection settings?')
except smtplib.SMTPServerDisconnected:
  print('Failed to connect to the server. Wrong user/password?')
except smtplib.SMTPException as e:
  print('SMTP error occurred: ' + str(e))
else:
  print('Sent')
```
