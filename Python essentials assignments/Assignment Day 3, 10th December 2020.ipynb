```python
# Use IF ELSE and ELIF to write a program in python for your Report Cards.

name = input('enter the percentage of student:')

#stripping zeroes and converting it into interger
a = int(name.strip())
if(a>85):#if percent > 85 --A
    print("Grade - 'A'")
elif(a<=85 and a>80):#80<percent80=85  --A-
    print("Grade - 'A-'")
elif(a>70 and a<=80):#70<percent<=80 --B
    print("Grade - 'B'")
elif(a>60 and a<=70):#60<percent<=70  --C
    print("Grade - 'C'")
elif(a>40 and a<=60):#40<percent<=60  --D
    print("Grade - 'D'")
elif(a<=40):#percent<=40  --E
    if(a<=35):#failed condition
        print("Grade - 'E' Candidate failed")
    else:
        print("Grade - 'E'")
```

```python
# Use For Loop to Print Prime Numbers in between 1 to 1000

for num in range(1,1000 + 1):  
   if num > 1:  
       for i in range(2,num):  
           if (num % i) == 0:  
               break  
       else:  
           print(num) 
```

```python
# Write a program for printing the tables from 1,10 using Nested For Loop 

for i in range(1, 11):
    print("i =", i, ":", end=" ")
    for j in range(1, 11):
        print("{:2d}".format(i * j), end=" ")
    print()
```

```python
# Write a program to Print X Prime Numbers using While Loop starting from 0, and take the INput of X from the user

# usedfor loop

X = int(input("Enter number: "))
for i in range(0,X):
  if i>-1:
    for j in range(2,i):
        if(i % j==0):
            break
    else:
        print(i)
```
